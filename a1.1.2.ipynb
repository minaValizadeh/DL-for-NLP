```python
import pandas as pd
import tensorflow as tf
from tensorflow import set_random_seed
from pylab import *
import matplotlib.pyplot as plt1
from sklearn import preprocessing
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

BATCH_SIZE = 1
SEED = 100
np.random.seed(SEED)
set_random_seed(SEED)



Cancer_data = load_breast_cancer()
Data = Cancer_data['data']
Label = Cancer_data['target']

# Normalizing data, all the features will be in range [0,1]
Data = preprocessing.normalize(Data)

# splitting data to train and test set
X, X_test, Y, y_test = train_test_split(Data, Label, test_size=0.3, random_state=SEED)
# splitting train to train and validation set
X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.1, random_state=SEED)

Cancer_data = load_breast_cancer()
Data = Cancer_data['data']
Label = Cancer_data['target']



def Cancer_model(x, y, lr):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(1, input_dim=30, activation='sigmoid'))
    adam = tf.keras.optimizers.Adam(learning_rate=lr)
    model.compile(optimizer=adam, loss='binary_crossentropy', metrics=['accuracy'])
    history = model.fit(x, y, epochs=30, validation_data=(X_val, y_val))


    training_loss = history.history['loss']
    validation_loss = history.history['val_loss']
    
    training_Acc = history.history['acc']
    validation_Acc = history.history['val_acc']
  # number of epochs
    epoch_count = range(1, len(training_loss) + 1)
    

  # Visualize loss history
    plt1.plot(epoch_count, training_loss, 'r--')
    plt1.plot(epoch_count, validation_loss, 'b-')
    plt1.legend(['Training Loss', 'Validation Loss'])
    plt1.xlabel('Epoch')
    plt1.ylabel('Loss')
    plt1.show();
  
  # Visualize accuracy history
    plt1.plot(epoch_count, training_Acc, 'r--')
    plt1.plot(epoch_count, validation_Acc, 'b-')
    plt1.legend(['Training Accuracy', 'Validation Accuracy'])
    plt1.xlabel('Epoch')
    plt1.ylabel('Accuracy')
    plt1.show();

    return model
```

```python
model1=Cancer_model(X_train,y_train,0.01)
```

```python
model2=Cancer_model(X_train,y_train,0.025)
```

```python
model5.evaluate(X_test,y_test,verbose=0)
```
